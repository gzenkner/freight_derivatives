```python
import pandas as pd
import os
from pathlib import Path
import json
```

```python
root_dir = Path("/Users/gabrielzenkner/projects/freight_derivatives/datasets/indices/json")
def load_bdi(bdi_path: str) -> pd.DataFrame:
    complete_paths = sorted(bdi_path.glob("*.json"))
    df_bdi = pd.DataFrame()

    for p in complete_paths:
        with open(p, 'r') as file:
            data = json.load(file)["data"]
        df = pd.json_normalize(data)
        df["IndexShortName"] = str(p.stem)
        df_bdi = pd.concat([df_bdi, df], ignore_index=True)
    return df_bdi
df_bdi = load_bdi(root_dir)
```
